```python
import pandas as pd

df = pd.read_csv("titanic.csv")

df.head()
```

```python
df["age"].notnull().value_counts() # Nombre de personnes ou l'age est présente et celle ou elle est nul
```

```python
df.groupby("survived").size() # Nombre de personnes qui ont survécu
```

```python
df[df["survived"] == 1]["age"].mean() # Age moyen des survivant
```

```python
df[df["survived"] == 1]["fare"].mean() # Prix moyen du ticket pour les survivants
```

```python
df.groupby("survived")["age"].mean() # Comparatif de la moyen d'age entre les survivants et les morts
```

```python
df.groupby(["survived","sex"]).size() # Comparatif de nombre de femme/homme qui ont survécu
```

```python
df.groupby(["survived"])["sex", "age", "pclass"].count() # Nombre de personne qui ont survécu ou non par rapport à certain attribut
```

```python
# ./sncf/stops.txt ( stop_name ) -> stop_id -> ./sncf/stop_times.txt
stop = pd.merge(stops,stop_times)
stop = stop [['stop_id','stop_name','trip_id','arrival_time','departure_time']]
stop.head()
```

```python
# ./sncf/stop_times.txt -(arrival, departure) -> trip_id ->  ./sncf/trips.txt
stop_trip = pd.merge(stop,trips)
stop_trip = stop_trip[['stop_id','stop_name','trip_id','arrival_time','departure_time','route_id']]
stop_trip.head()
```

```python
# ./sncf/trips.txt -> route_id -> ./sncf/routes.txt
stop_trip_route = pd.merge(stop_trip,routes)
stop_trip_route = stop_trip_route[['stop_id','stop_name','trip_id','arrival_time','departure_time','route_id','route_long_name']]
stop_trip_route.head(100)
```

```python
# ./sncf/routes.txt (route_long_name)
print(stop_trip_route["stop_name"].unique())
```

```python
gare_est_paris = stop_trip_route.loc[stop_trip_route['stop_name'] == 'Gare de Paris-Est']
gare_est_paris.dropna()
gare_est_paris
```
